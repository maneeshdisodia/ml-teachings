```python
purpose=pd.get_dummies(loan['purpose'],drop_first=True)
```

```python
loan.drop('purpose',axis=1,inplace=True)
```

```python
loan =pd.concat([loan,purpose],axis=1)
```

```python
loan
```

```python
#train split
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=loan.drop('not.fully.paid',axis=1)
```

```python
y=loan['not.fully.paid']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
#make model object
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree= DecisionTreeClassifier()
```

```python
dtree.fit(X_train,y_train)
```

```python
pred=dtree.predict(X_test)
```

```python
#evaluation
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
print(confusion_matrix(y_test,pred))
print('\n')
print(classification_report(y_test,pred))
```

# random decesion tree

```python
rfc=RandomForestClassifier(n_estimators=16)
```
