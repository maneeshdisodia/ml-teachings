```python
x=df[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms',
       'Avg. Area Number of Bedrooms', 'Area Population']]
```

```python
y=df['Price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.4, random_state=101)
```

```python
X_train.info()
```

```python
lm=LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
print(lm.intercept_)
```

```python
lm.coef_
```

```python
x.columns
```

```python
pd.DataFrame(data=lm.coef_,index=x.columns,columns=['coeff'])
```

# Predictions

```python
prediction = lm.predict(X_test)
```

```python
prediction
```

```python
sns.distplot(prediction - y_test)
```

```python
plt.scatter(y_test,prediction)
```

```python
from sklearn import metrics
```

```python
metrics.mean_absolute_error(y_test,prediction)
```

```python
metrics.mean_squared_error(y_test,prediction)
```

```python
np.sqrt(metrics.mean_squared_error(y_test,prediction))
```

# Bosten prediction

```python
from sklearn.datasets import load_boston
```

```python
boston=load_boston()
```

```python
boston.keys()
```

```python
print(boston.feature_names)
```

```python
print(boston.data[0])
```

```python
bostanData=pd.DataFrame(data=boston.data,columns=boston.feature_names)
```

```python
bostanData.head(2)
```

```python
sns.pairplot(bostanData)
```

```python
B_train,B_test,B_label_train,B_label_test=train_test_split(bostanData,boston.target,test_size=0.4, random_state=101)
```

```python
lm2=LinearRegression()
```

```python
lm2.fit(B_train,B_label_train,sample_weight=0.9)
```

```python
lm2.intercept_
```

```python
lm2.coef_
```

```python
boston.feature_names
```

```python
boston_coeff=pd.DataFrame(data=lm2.coef_,index=boston.feature_names,columns=['coeff'])
```

```python
boston_coeff
```

```python
bostanData.head(2)
```

```python
boston_predict=lm2.predict(B_test)
```

```python
plt.scatter(B_label_test,boston_predict)
```

```python
sns.distplot((boston_predict - B_label_test))
```

# boston evaluation

```python
metrics.mean_absolute_error(B_label_test,boston_predict)
```

```python
metrics.mean_squared_error(B_label_test,boston_predict)
```

```python
np.sqrt(metrics.mean_squared_error(B_label_test,boston_predict))
```
