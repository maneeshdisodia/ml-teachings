```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x= np.linspace(0,11,10)
x
```

```python
y=x**2
```

```python
a,b= plt.subplots(1,2)
b[0].plot(x,y,label='1st one')
b[1].plot(y,x,label='2nd one')
a.tight_layout()
a.legend()
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(y**2,x,color='g',lw=2,ls='--')
axes.set_xlim([0,40])
axes.set_ylim([0,40])
```
