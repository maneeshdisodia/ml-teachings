```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x= np.linspace(0,5,11)
```

```python
y=x**2
```

```python
#functional
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('xy plot title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#object oriented approch
x,y
```

```python
fig=plt.figure()
axes= fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('x lable')
axes.set_ylabel('y lable')
axes.set_title('ttle')
```

```python
figg=plt.figure()
axes1=figg.add_axes([0.1,0.1,0.8,0.8])
axes2=figg.add_axes([0.2,0.4,0.5,0.3])
axes1.plot(x,y)
axes1.set_xlabel('axes1-X')
axes1.set_ylabel('axes1-Y')
axes1.set_title('larger')
axes2.plot(y,x)
axes2.set_xlabel('axes2-X')
axes2.set_ylabel('axes2-Y')
axes2.set_title('smaller')
```

```python
c,d=plt.subplots(nrows=1,ncols=2)
d[0].plot(x,y)
d[0].set_title('plot1')

d[1].plot(y,x)
d[1].set_title('plot2')

#c.tight_layout()
```

```python
###figure Size and control
```

```python
fig=plt.figure(figsize=(8,2))
ax= fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(y**2,x)
```

```python
aa,bb=plt.subplots(2,1,figsize=(11,2))
bb[0].plot(x,y)
bb[1].plot(y,x)
aa.tight_layout()
```

```python
aa.savefig('mypng')
```

```python
fig=plt.figure()
ax =fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='x-y')
ax.plot(x,x**3,label='y-x')
ax.legend()
```
